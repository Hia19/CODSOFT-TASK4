---


\
Since with graphical representation it is prominent that the variable "TV" is the most correlated with "Sales", hence performing Linear Regression Model using the variable "TV".


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(adv['TV'], adv['Sales'], test_size=0.3,random_state=0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(pd.DataFrame(X_train), y_train)
```

```python
res= model.predict(pd.DataFrame(X_test))
print(res)
```

```python
model.coef_
```
